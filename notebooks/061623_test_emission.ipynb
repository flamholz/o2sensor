```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
spectra = pd.read_excel('../data/061623_PtF84/061623_emission_spectra.xlsx')
spectra.head(2)
```

```python
cols = ['acetone', 'PtF84_acetone', 'PtPFOB']
labels = 'acetone blank,PtF84 in acetone,PtPFOB'.split(',')

for col, label in zip(cols, labels):
    plt.plot(spectra['wavelength'], spectra[col], label=label)
plt.legend()
plt.xlabel('Wavelength (nm)')
plt.ylabel('Emission')
plt.tight_layout()
plt.savefig('../output/061623_emission_spectra.png', dpi=300, facecolor='w')
plt.show()
```
